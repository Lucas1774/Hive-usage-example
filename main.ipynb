3 - Crear vista para cada país nombre, código, esperanza de vida en 2018 y inversión en 2018

```python
%%writefile createcreateviewscript.hql
use paisesdb;

CREATE VIEW vistaapartado3 AS
SELECT a.Country_Name AS Pais, a.Country_Code AS Codigo,
a.year2018 AS esperanza_en_2018, b.year2018 AS inversion_en_2018
FROM expectancy a, inversion b
WHERE a.Country_Code = b. Country_Code;
```

4 - c los cinco paises con mayor inversión en investigación en 2010, usando limit 5 y order by

```python
%%writefile createqueryejercicio4cscript.hql
use paisesdb;

SELECT a.Country_Name as Pais, a.year2010 as Inversion_2010
FROM inversion a
WHERE a.year2010 IS NOT NULL
ORDER BY Inversion_2010 DESC
LIMIT 5
```
